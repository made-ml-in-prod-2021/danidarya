```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import pathlib
```

```python
path_dir = str(pathlib.Path('__file__').parent.absolute())
data_path = path_dir.replace('notebooks','data_raw/')+'heart.csv'
data = pd.read_csv(data_path)
```

```python
sns.histplot(data['target'],color='mediumseagreen')
```

```python
numerical = ['age','trestbps','chol','thalach','oldpeak']
cathegorial = ['sex','cp','fbs','restecg','exang','slope','ca','thal']
for feature in cathegorial:
    plt.figure(figsize=(12, 3))
    sns.countplot(x=data[feature], hue=data['target'],palette='viridis')
    plt.show()
```

```python
corrmat = data.corr()
f, ax = plt.subplots(figsize=(14,10))
sns.heatmap(corrmat,cmap=plt.cm.viridis,linecolor='white', annot=True)   
```

```python
sns.pairplot(data[numerical + ['target']], hue='target',palette="viridis")
```
